# 2. Load Model from Hugging Face

Download the trained language detection model.

```python
from huggingface_hub import snapshot_download

print("d1")

repo_path = snapshot_download(
    repo_id="sepzyd/language-identification-nb"
)

print("d3")
print("Model downloaded to:", repo_path)
```

```python
import joblib
import json
import pandas as pd
from pathlib import Path

repo = Path(repo_path)

model = joblib.load(repo / "nb_model.joblib")
vectorizer = joblib.load(repo / "vectorizer.joblib")

with open(repo / "label_mapping.json") as f:
    label_map = json.load(f)

df = pd.read_csv(repo / "language_codes.csv")

code_to_language = dict(zip(df.code, df.language))

print("Model loaded successfully")
```

# 3. Language Detection Demo

Type a sentence and the system will detect the language.

```python
def detect_language(text):

    X = vectorizer.transform([text])

    pred = model.predict(X)[0]

    iso_code = label_map[str(pred)]

    language = code_to_language.get(iso_code, iso_code)

    return language
```

```python
while True:

    text = input("Enter text (type 'quit' to stop): ")

    if text == "quit":
        break

    if text.strip() == "":
        print("Please enter text")
        continue

    print("Detected language:", detect_language(text))
```
